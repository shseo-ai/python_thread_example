```python
import threading
import time

lock = threading.Lock() # lock 객체
shared_num = 0 # 다루고자 하는 데이터
flag = False # 대기 플래그

class ThreadTest1(threading.Thread):
    def run(self):
        global lock
        global shared_num
        global flag
        exit_flag = True # 대기 뿐 아니라 스레드를 종료하기 위한 플래그

        for i in range(1000):
            lock.acquire() # lock 실행
            if flag == False: # 대기 플래그 체크
                shared_num += 1
                if shared_num == 500: # 대기 조건 체크
                    flag = True # 대기 플래그 활성화 -> 다른 쓰레드에서도 작업 중지 가능
                    print('\n', '#'*50)
                    print("T1 make result: ",  shared_num)
                    print('\n', '#'*50)

                    #exit flag를 이용하여 값을 얻어낸 스레드에서 작업을 추가적으로 진행가능
                    # exit_flag = False
                else:
                    print(threading.currentThread().getName(), ': ', shared_num)            
            else: #대기 플래그가 활성화 되었을 시                                                                                                                                                    
                print('T1 is stoped...')
                lock.release() # lock 해제
                if exit_flag == True:
                    return None
                else: 
                    time.sleep(10)
            lock.release() # lock 해제

class ThreadTest2(threading.Thread):
    def run(self):
        global lock
        global shared_num
        global flag
        exit_flag = True # 대기 뿐 아니라 스레드를 종료하기 위한 플래그

        for i in range(1000):
            lock.acquire() # lock 실행
            if flag == False: #플래그 체크
                shared_num += 1
                if shared_num == 500: # 대기 조건 체크
                    flag = True # 대기 플래그 활성화 -> 다른 쓰레드에서도 작업 중지 가능
                    print('\n', '#'*50)
                    print("T2 make result: ",  shared_num)
                    print('\n', '#'*50)

                    #exit flag를 이용하여 값을 얻어낸 스레드에서 작업을 추가적으로 진행가능
                    # exit_flag = False
                else:
                    print(threading.currentThread().getName(), ': ', shared_num)            
            else: #대기 플래그가 활성화 되었을 시                                                                                                                                                    
                print('T2 is stoped...')
                lock.release() # lock 해제
                if exit_flag == True:
                    return None
                else: 
                    time.sleep(10)
            lock.release() # lock 해제

if __name__ == "__main__":
    t1 = ThreadTest1(name="Thread1")
    t2 = ThreadTest2(name="Thread2")

    t1.start()
    t2.start()
```
